### Data Science London + Scikit-learn


```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
```

```python
path = r'C:\Users\user\Documents\GitHub\2nd-ML100Days\homework\Kaggle\data-science-london-scikit-learn'
trainX = r'C:\Users\user\Documents\GitHub\2nd-ML100Days\homework\Kaggle\data-science-london-scikit-learn\train.csv'
trainY = r'C:/Users/user/Documents/GitHub/2nd-ML100Days/homework/Kaggle/data-science-london-scikit-learn/trainLabels.csv'
testX = r'C:/Users/user/Documents/GitHub/2nd-ML100Days/homework/Kaggle/data-science-london-scikit-learn/test.csv'
x = pd.read_csv(trainX, header=None)
y = pd.read_csv(trainY, header=None)
test_x = pd.read_csv(testX, header=None)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y[0], test_size=0.5, random_state=4)

clf = GradientBoostingClassifier(n_estimators=100, max_depth =5, random_state=7)

# 先看看使用預設參數得到的結果
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred = clf.predict(test_x)
Id = np.arange(1,9001)
```

```python
result = pd.DataFrame({'Id': Id, 'Solution': y_pred}, columns=['Id', 'Solution'])
result.to_csv(path+'/'+"result.csv",index=False)
```
